```python
# Import all libraries required
import pandas as pd
import numpy as np
```

```python
# Data Ingestion
df = pd.read_csv('ds_salaries.csv')
df.head()
```

```python
# Finding number of rows and columns in dataset df
df.shape
```

```python
# Extracting eneral Information of the dataset
df.info()
```

```python
# Checking for the missing values
df.isna().sum()
```

```python
# Mentioning full forms of : 'SE', 'MI', 'EN', 'EX'
df['experience_level'].replace(['SE', 'MI', 'EN', 'EX'],
                               ['Senior_level','Mid_level','Entry_level','Executive_level'],
                               inplace=True)
df.head()
```

```python
# Mentioning full forms of : 'FT', 'CT', 'FL', 'PT'
df['employment_type'].replace(['FT', 'CT', 'FL', 'PT'],['Full-time','Contract','Freelancer','Part-time'], inplace=True)
df.head()
```

```python
# Finding conversion rates of salary recieved by employees
df['conversion_rates'] = df['salary']/df['salary_in_usd']
df.head()
```

```python
# Converting country codes of employee_residence column in proper one
import country_converter as coco
df['employee_residence'] = coco.convert(names=df['employee_residence'])
df.head()
```

```python
# Converting country codes of company_location column
import country_converter as coco
df['company_location'] = coco.convert(names=df['company_location'])
df.head()
```

```python
# Finding foregin employees
df['foreign_employee']=np.where(df['employee_residence']==df['company_location'],"NO","YES")
df.head()
```

```python
# Renaming remote ratio elements
df['remote_ratio'].replace([0,50,100],['On-Site','Hybrid','Remote'], inplace=True)
df.rename(columns={'remote_ratio':'job_type'}, inplace=True) #Renaming the column
df.head()
```

```python
# Mentioning full forms of : 'L', 'M', 'S'
df['company_size'].replace(['L', 'M', 'S'],
                               ['Large','Medium','Small'],
                               inplace=True)
df.head()
```

```python
# Saving cleaned and transformed data 
df.to_csv('data_science_salary_details.csv', index=False)
```
